```python
# Entrega
data_entrega = input(f"Para qual data você gostaria? ")
endereço_entrega = input(f"Em qual endereço você gostaria de receber?")
print("A entrega será feita no dia {data_entrega}, no endereço {endereço_entrega}")
```

```python
# Tipo de chocolate
tipo_chocolate_usuario = int(input("Qual tipo de chocolate você gostaria? "))
list = ["1.chocolate ao leite, 2.chocolate meio amargo, 3.chocolate branco"]
if tipo_chocolate_usuario == 1:
  print("Você escolheu chocolate ao leite.")
elif tipo_chocolate_usuario == 2:
    print("Você escolheu chocolate meio amargo.")
elif tipo_chocolate_usuario == 3:
      print("Você escolheu chocolate branco.")
else:
        print("opção inválida")
```

```python
# Peso e quantidade
quantidade_usuario = int( input("quantos ovos você gostaria? "))
gramas_usuario = int( input("quantas gramas cada ovo terà? "))
total_gramas = gramas_usuario * quantidade_usuario
total_Kg = total_gramas / 1000
preço_kg = 115
preço_total = preço_kg * total_Kg
```

```python

# Embalagem Personalizada
embalagem_especial = input(f"Você gostaria de uma embalagem personalizada? (S/N)")
if embalagem_especial == "Sim":
  preço_total += 20
```

```python
# Desconto
if quantidade_usuario >= 8:
    print("Parabéns você ganhou 5% de desconto na sua compra!")
    preço_total *= 0.05 # Applying the 5% discount by multiplying by 0.95
```

```python
# resumo final
print(f"Resumo final!")
print(f"quantidade de ovos: {quantidade_usuario}")
print(f"gramas por ovo: {gramas_usuario}")
print(f"total de gramas: {total_gramas}")
print(f" peso total: {total_gramas}g {total_Kg:.2f} kg")
print(f"Valor total da encomenda: R$ {preço_total:.2F}")
print(f"A encomenda será entregue no endereço: {endereço_entrega} na data {data_entrega}")
```
